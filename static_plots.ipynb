# Plot static stuff for the site

```python
import json
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams["figure.dpi"] = 200
plt.rcParams["font.family"] = "Inter"
```

## Histograms

```python
with open("./data/all_hist_stats.json") as fin:
    stuff_to_track = json.load(fin)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=len(stuff_to_track), figsize=(28, 4))

for i, var in enumerate(stuff_to_track):
    axs[i].hist(stuff_to_track[var])
    axs[i].set_title(var)
```

```python
# Summary stats
df = pd.DataFrame.from_dict(stuff_to_track)
```
